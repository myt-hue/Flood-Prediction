```python
# Importing Relevant Functions
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 1000)
from datetime import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('IndianFloodInventroy(IFI).csv')
```

```python

# Replace 'nan' with actual NaN values
df_description = df['Description of Casualties/injured'].replace('nan', np.nan)
# Remove NaN values from the 'Description of Casualties/injured' column and 'Extent of Damage' column 
df_description = df.dropna(subset=['Description of Casualties/injured', df.columns[17]])
```

```python
df_description.reset_index()
```

```python
# Getting Rid of Columns that are not useful for our analysis
df = df.drop(columns = ['Human fatality', 'Human injured', 'Animal Fatality', 'Event Souce ID'])
df
```

```python
df['Human Displaced'].fillna(0)
```

```python

result_df = df.groupby(['Main Cause'])['Human Displaced'].sum().reset_index()
result_df['Main Cause'] = df['Main Cause'].replace({'Heavy rain': 'Heavy Rain'})
print(result_df)
```

```python
# Removing Duplicates
result_df_combined = result_df.drop_duplicates().groupby(['Main Cause'])['Human Displaced'].sum().reset_index()
result_df_combined
```

```python
# Create a bar plot using Seaborn
plt.figure(figsize=(10, 6))
filtered_df = result_df_combined[result_df_combined['Main Cause'].isin(['Heavy Rain', 'Monsoonal rain'])]
sns.barplot(x='Main Cause', y='Human Displaced', data = filtered_df, palette='viridis')
# Set labels and title
plt.xlabel('Main Cause')
plt.ylabel('Humans Displaced')
plt.title('Causes of Human Displacement')

# Show the plot
plt.show()
```

```python
# Geomap of the location of flood using coordintes
import geopandas
import matplotlib.pyplot as plt


# Create GeoDataFrame
gdf = geopandas.GeoDataFrame(
    df, geometry=geopandas.points_from_xy(df.Longitude, df.Latitude))

# Set the coordinate reference system (CRS)
crs_info = 'GEOGCS["GCS_WGS_1984",DATUM["D_WGS_1984",SPHEROID["WGS_1984",6378137.0,298.257223563]],PRIMEM["Greenwich",0.0],UNIT["Degree",0.0174532925199433]]'
gdf.crs = crs_info

# Assigning 'Main Cause'
heavy_rain = gdf[gdf['Main Cause'] == 'Heavy Rain']
monsoonal_rain = gdf[gdf['Main Cause'] == 'Monsoonal rain']
# Load world shapefile
world_shapefile_path = 'india_flood_inventory_corrected.shp'
world = geopandas.read_file(world_shapefile_path)

# Set the coordinate reference system (CRS) for the world GeoDataFrame
world.crs = crs_info


fig, ax = plt.subplots(figsize=(15, 15))
base = world.plot(color='white', edgecolor='gray', ax=ax)
# Scatter plots for each category
ax.scatter(heavy_rain['Longitude'], heavy_rain['Latitude'], c='darkblue', alpha=1, s=3, label='Heavy Rain')
ax.scatter(monsoonal_rain['Longitude'], monsoonal_rain['Latitude'], c='orange', alpha=1, s=3, label='Monsoonal Rain')


# Adding labels and title
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Flood Status in India')

# Adding legend
ax.legend(markerscale = 3, loc='upper right')

plt.show()
```

```python
# Furthur Focus on Locations affected by Monsoonal Rains, but adjusting the size of Area Affected 

# Filtering by Monsoonal Rains 
monsoonal_rain = gdf[gdf['Main Cause'] == 'Monsoonal rain']
fig, ax = plt.subplots(figsize=(15, 15))
base = world.plot(color='white', edgecolor='gray', ax=ax)

# Plot for 'Monsoonal' rains with marker size based on 'Area Affected'
# Normalize 'Area Affected' for visualization
max_height = monsoonal_rain['Area Affected'].max()
marker_size = (monsoonal_rain['Area Affected'] / max_height) * 100

ax.scatter(monsoonal_rain['Longitude'], monsoonal_rain['Latitude'], c='blue', alpha=0.5, s=marker_size, label='Monsoonal Rain')


# Adding labels and title
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Monsoonal Rain in India')

# Adding legend
ax.legend(markerscale=3, loc='upper right')

plt.show()
```

The word clouds below give us a better understanding of how to narrow down our analysis. "Persons died" is the most frequent word from the first word cloud. "Districts" and "Damage" are the most frequent words from the second word cloud.

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.colors import to_rgba

# Convert elements to strings
df_description['Description of Casualties/injured'] = df_description['Description of Casualties/injured'].astype(str)
text = ' '.join(df_description['Description of Casualties/injured'])

# Generate a word cloud with different colors based on 'Flood Type'
wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='viridis').generate(text)

# Create a custom color function
color_dict = {label: to_rgba(color) for label, color in zip(df_description['Main Cause'].unique(), plt.cm.viridis(range(len(df_description['Main Cause'].unique()))))}

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8,8))

# Plotting word cloud
ax.imshow(wordcloud, interpolation='bilinear')
ax.axis('off')
ax.set_title('Word Cloud for Description of Fatalities/Injuries (Colored by Flood Status)', pad=20)

# Annotate with Flood Types and create legend patches
specific_rows = [0, 1979]
legend_patches = []
for index, row in df_description.iloc[specific_rows].iterrows():
    label = row['Main Cause']
    color = color_dict.get(label, 'black')  # Get the color associated with the label
    legend_patches.append(mpatches.Patch(color=color, label=label))
    ax.annotate(label, xy=(0.5, 0.5), xytext=(index, index), ha='center', va='center',
                bbox=dict(boxstyle='round', alpha=0.1, facecolor='white'))

# Add legend on the side
ax.legend(handles=legend_patches,bbox_to_anchor=(0.5, -0.2), loc= 'upper center', title="Flood Satus", borderaxespad=0.1,prop={'size': 8})

plt.show()

```
